```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
air = pd.read_csv('../data/AirData.csv')
air.head()
```

# adding a column to extract year

```python
air['year'] = pd.DatetimeIndex(air['datemon']).year
air['year']=air['year'].astype(int)
air.head()
```

# Counting no of cities

```python
print("Total cities", len(air['City'].unique()))
air['City'].unique().tolist()
```

Observations:

PM2.5, O3 are usually high throughout the year
CO has been found consistently low over the years


# 1.Univariate Analysis

```python
pollutants = ['PM2.5','NO','NO2','CO','SO2','O3']
df = air.set_index('datemon')
axes = df[pollutants].plot(marker='.',alpha = 0.5, linestyle = 'None', figsize = (16,20), subplots = True)
for ax in axes:
    ax.set_xlabel('Years')
    ax.set_ylabel('ug/m3')
```

```python
plt.title('Distribution of AQI Buckets 2018-2020')
sns.countplot(x='AQI Bucket', data=air, hue='year')
```

Overall we see we have a large number of Moderate and Satisfactory bucket counts which is a good thing and low number of Poor and Severe buckets over 2018-2020.
If we observe carefully, 2019 has highest Moderate and Satisfactory buckets which can be explained by lower pollution levels by the lockdown.
And as compared to the previous years 2018 & 2019, for all the buckets we see a steep decline in 2020, and in case of "Good" bucket - higher number than previous years which can be seen as a positive sign.

# Year wise break up

```python
y18=air[air['year']==2018]
y18.head()
plt.title('Distribution of City level AQI 2018')
sns.barplot(x='AQI', y='City', hue='year', ci=None, data=y18)
```

```python
y19=air[air['year']==2019]
y19.head()
plt.title('Distribution of City level AQI 2019')
sns.barplot(x='AQI', y='City', hue='year', ci=None, data=y19)
```

```python
y20=air[air['year']==2020]
y20.head()
plt.title('Distribution of City level AQI 2020')
sns.barplot(x='AQI', y='City', hue='year', ci=None, data=y20)
```

From detailed year wise analysis we concluded that Aizwar has been recorded from 2020 which was missing for previous 2 years.It may result in variations while calculting the average of each year & each city.
Ahmedabad is always on top in AQI resulting very poor air quality. It might indicate that due to heavy industrialization.
2nd highest will be Delhi as resulting a high average AQI.

```python
## AQI
print("Minimum and maximum range of AQI: {air.AQI.min()} - {air.AQI.max()}")
plt.title('AQI levels over 2018 to 2020')
sns.lineplot(x='year',y='AQI', data=air);
```
